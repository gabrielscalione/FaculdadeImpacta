# Clusterização k-Means (ou k-Médias)

O algoritmo k-Means busca por um determinado número de clusters dentro de um dataset multidimensional não rotulado.

Faz duas hipóteses:

- O "centro do cluster" é a média aritmética de todos os pontos pertencentes ao cluster.
- Cada ponto é mais próximo de seu centro de cluster que outros centros.



```
from sklearn.datasets import make_blobs

X, y_true = make_blobs(n_samples=300,
                       centers=4,
                       cluster_std=0.6,
                       random_state=0)
```

Padrão dos estimadores (modelos) do sklearn:

- Ajustar modelo aos dados.

```
from sklearn.cluster import KMeans 
```

```
kmeans = KMeans(n_clusters =4)
```

```
y_kmeans = kmeans.predict(X)
```

# Algoritmo k-Means: Expectation-Maximization

1. "Chute" alguns centros de clusters
2. Repita até a convergência:

  2.1 E-step: atribua pontos ao centro de cluster mais próximo.
  
  2.2 M-step: defina os centros de clusters para a média

Cada repetição dos passos E e M irão sempre resultar numa melhor estimativa das características dos clusters.

# Ressalvas do algoritmo

## 1. O resultado ótimo global pode não ser atingido.

image.png

Fonte: https://jakevdp.github.io/PythonDataScienceHandbook/05.11-k-means.html

O resultado da figura acima foi gerada por uma implementação ingênua (naïve) do KMeans.

As bibliotecas implementam técnicas para tentar fugir dessa situação. Por exemplo, rodar o algoritmo para vários "chutes" iniciais. Parâmetro `n_init` do sklearn.

## 2. O número de clusters deve ser decidido antes de rodar o algoritmo.

```
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centros = kmeans.cluster_centers_
fig, ax = plt.subplots(figsize=(12,6))

ax.scatter(X[:, 0], X[:, 1],
           c=y_kmeans,
           s=50,
           cmap='viridis');

ax.scatter(centros[:, 0],
           centros[:, 1],
           c='black',
           s=200,
           alpha=0.5);
```

```
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centros = kmeans.cluster_centers_
fig, ax = plt.subplots(figsize=(12,6))

ax.scatter(X[:, 0], X[:, 1],
           c=y_kmeans,
           s=50,
           cmap='viridis');

ax.scatter(centros[:, 0],
           centros[:, 1],
           c='black',
           s=200,
           alpha=0.5);
```

```
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=15)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centros = kmeans.cluster_centers_
fig, ax = plt.subplots(figsize=(12,6))

ax.scatter(X[:, 0], X[:, 1],
           c=y_kmeans,
           s=50,
           cmap='viridis');

ax.scatter(centros[:, 0],
           centros[:, 1],
           c='black',
           s=200,
           alpha=0.5);
```

Quando o resultado é significativo?

## 3. É limitado a fronteiras (boundaries) **lineares**

```
from sklearn.datasets import make_moons
X, y = make_moons(200, noise=.05, random_state=0)
kmeans = KMeans(2, random_state=0)
kmeans.fit(X)
rotulos = kmeans.predict(X)
fig, ax = plt.subplots(figsize=(12,6))
centros = kmeans.cluster_centers_
ax.scatter(centros[:, 0],
           centros[:, 1],
           c='black',
           s=200,
           alpha=0.5)
ax.scatter(X[:, 0], X[:, 1],
           c=rotulos,
           s=50,
           cmap='viridis');
```

Alternativa: usar versão que faz uma transformação de kernel para projetar os dados em uma dimensão onde a separação linear é possivel.

Uma versão do k-Means com transformação de kernel é o estimador `SpectralClustering`.

```
from sklearn.cluster import SpectralClustering
modelo = SpectralClustering(n_clusters=2,
                            affinity='nearest_neighbors',
                            assign_labels='kmeans')
rotulos = modelo.fit_predict(X) # Não tem predict sozinho...
fig, ax = plt.subplots(figsize=(12,6))
ax.scatter(X[:, 0],
            X[:, 1],
            c=rotulos,
            s=50,
            cmap='viridis');
```
