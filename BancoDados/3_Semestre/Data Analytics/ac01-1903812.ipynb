A maioria dos métodos não altera o objeto que está sendo manipulado e sim devolve uma visão com aquela operação aplicada.

###Respondendo perguntas

Com os tipos de dados corrigidos, e nomes de colunas no padrão, podemos começar a responder algumas perguntas.

Para isso, podemos realizar seleção de linhas.

Uma das formas é usando o método `query`.

Quais são os estados que tem mais de 5 milhões de habitantes?

```
populacao.query('populacao > 5000000')
```

Quais são os estados que tem entre 1 milhão e 5 milhões?

```
populacao['populacao'] > 5000000
```

```
(populacao['populacao'] >= 1000000) & (populacao['populacao'] <= 5000000)
```

```
populacao['porcentagem_populacao'] < 0.5
```

E usar esse vetor no dataframe para devolver aquelas posições nas quais os valores são verdadeiros:

Junção dos dataframes `populacao` e `area`.

Método `merge`, estilo join do SQL.

Parâmetros principais:

- `right` (primeiro parâmetro): Qual dataframe da direita.
- `how`: default: inner. Opções: {‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’}.
- `on`: nome da coluna de join (se forem as mesmas nos dois dataframes).
- `left_on`: nome da coluna de join da esquerda (se forem difrentes nos dois dataframes)
- `right_on`: nome da coluna de join da direita (se forem difrentes nos dois dataframes)


```
populacao.merge(area, on='uf')
```

Novamente, o Dataframe original não é alterado.

Vamos armazenar o resultado em uma nova variável.

```
populacao_area = populacao.merge(area, on='uf')
```

Podemos adicionar novas variáveis, como a proporção da população em relação a área.

Para tanto, realizaremos as operações sobre as colunas, novamente de forma vetorizada.

Esse modo de realizar operações é bem mais rápido que usando um for.

E a sintaxe é a de operações normais.

```
populacao_area['populacao'] / populacao_area['area']
```

Para armazenar em uma nova coluna, basta atribuir com um nome que não exista.

```
populacao_area['densidade_populacional'] = populacao_area['populacao'] / populacao_area['area']
```
