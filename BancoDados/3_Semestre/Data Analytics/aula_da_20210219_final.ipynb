JunÃ§Ã£o dos dataframes `populacao` e `area`.

MÃ©todo `merge`, estilo join do SQL.

ParÃ¢metros principais:

- `right` (primeiro parÃ¢metro): Qual dataframe da direita.
- `how`: default: inner. OpÃ§Ãµes: {âleftâ, ârightâ, âouterâ, âinnerâ, âcrossâ}.
- `on`: nome da coluna de join (se forem as mesmas nos dois dataframes).
- `left_on`: nome da coluna de join da esquerda (se forem difrentes nos dois dataframes)
- `right_on`: nome da coluna de join da direita (se forem difrentes nos dois dataframes)


```
populacao.merge(area, on='uf')
```
