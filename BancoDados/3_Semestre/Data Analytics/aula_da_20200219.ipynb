E qual tipo de dado é essa tabela?

É um dos dois tipos principais utilizadas pela ferramenta Pandas, que é um **Dataframe**.


Uma Dataframe contém é uma estrutura de dados tabular de duas dimensões, heterogênea (as colunas podem ser de tipos diferentes), de tamanho mutável.

A estrutura contém **axes** (ou eixos) que são as linhas e colunas.

Tanto linhas quanto as colunas possuem índices. Esses índices podem ser de vários tipos, como inteiros e strings.

**Obs.:**: Pandas usa a biblioteca NumPy.

```
type(populacao)
```

Ver número de linhas e colunas

```
populacao.shape
```

Olhar o índice das colunas (Nesse exemplo são strings)

```
populacao.columns
```

###Segunda opção

Usarmos o método drop.

A maioria dos métodos não altera o objeto que está sendo manipulado e sim devolve uma visão com aquela operação aplicada.

Como um Dataframe tem duas dimensões, vários métodos podem ser aplicados nas linhas (axis=0, e padrão) ou nas colunas (axis=1).

No nosso caso, como estamos querendo remover uma coluna, precisamos passar o parâmetro axis com valor 1. (axis=1)

```
populacao.drop('Posição', axis=1)
```

Essa operação não foi aplicada no Dataframe original (retorno[0]) e nem atribuida a nenhuma variável.

Se quisermos manter o Dataframe original e tabalhar com a visão, podemos atribuir esse resultado a uma variável. No caso, iremos sobrescrever a variável populacao.

```
populacao = populacao.drop('Posição', axis=1)
```

```
populacao
```

```
retorno[0]
```

Nessa abordagem, não alteramos o Dataframe original (retorno[0]) e estamos trabalhando com uma visão dele na variável populacao.

Caso desejemos remover do dataframe original, passamos um parâmetro inplace=True isso, como nas células abaixo:

```
#Voltando a variável população com todas as colunas
populacao = retorno[0]
```

```
# A coluna Posição está lá
populacao.columns
```

```
# Usando o drop com o parâmetro inplace = True
populacao.drop('Posição', axis=1, inplace=True)
```

Com isso a coluna é removida do objeto original e não pode mais ser acessada por ninguém que está o referenciando.

```
populacao.columns
```

```
retorno[0].columns
```

###Identificando o problema

As colunas População e % da pop. total deveriam ser dos tipos int e float, respectivamente.

Foram carregadas como string por que na fonte de dados existem caracteres (lembrando que tabela html não tem tipo de dados e sim strings) que não permitem que o Pandas reconheça como numéricas.

No caso de População existe espaço separando milhares e na porcentagem existe vírgula separando as casas decimais e um caracter de % ao final.

Para podermos converter esses tipos, precisamos limpar essas strings, tirando esses caracteres.

Para população: remover espaços antes de converter.




```
'esta é uma string com espaços'.replace(' ', '')
```

```
populacao['População'][0].replace(' ', '')
```

```
int(populacao['População'][0].replace(' ', ''))
```

Após testarmos a limpeza com um elemento e vendo como fazer isso no Python (células acima), aplicamos para toda a coluna de forma vetorizada.


```
populacao['População'].str.replace(' ', '')
```

Para converter o tipo de dado, usamos astype.

```
populacao['População'].str.replace(' ', '').astype(int)
```

Após testarmos a aplicação (lembrando que nada foi alterado até então), fazemos a atribuição do resultado na coluna desejada.

```
populacao['População'] = populacao['População'].str.replace(' ', '').astype(int)
```

Parâmetros principais:

- `right` (primeiro parâmetro): Qual dataframe da direita.
- `how`: default: inner. Opções: {‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’}.
- `on`: nome da coluna de join (se forem as mesmas nos dois dataframes).
- `left_on`: nome da coluna de join da esquerda (se forem difrentes nos dois dataframes)
- `right_on`: nome da coluna de join da direita (se forem difrentes nos dois dataframes)
